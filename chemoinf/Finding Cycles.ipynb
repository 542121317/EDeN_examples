## EDeN Cycle Annotator

we will get to see how to find cycles in molecules using EDeN functions 

```python
# setting up data grabber
%matplotlib inline
from eden.converter.graph.gspan import gspan_to_eden
from itertools import islice

def get_graphs(size=5):
    # returns a generator over nx.graph
    return  islice(gspan_to_eden('http://www.bioinf.uni-freiburg.de/~costa/bursi_pos_train.gspan'),size)
```


Annotated graphs will have the lists part_name and part_id added to the nodes.

Below I show the part_name (which is human readable) on the graphs.

'part_name's tell you which atoms are in the cycle.

'part_id' is a unique id for each cycle, eg all nodes belonging to the same cycle will have the same part_id.

```python
from eden.modifier.graph.cycle import Annotator
from eden.util.display import draw_graph_set
#getting graphs
graphs=get_graphs(4)

#creating an annotator
cycle_annotator = Annotator()
annotated_graphs = cycle_annotator.transform(graphs)

# draw annoated graphs
draw_graph_set(annotated_graphs,n_graphs_per_line=2, size=10,
                       colormap='Paired', invert_colormap=False,node_border=0.5,vertex_label='part_name',
                       secondary_vertex_label='part_id',
                       vertex_alpha=0.5, edge_alpha=0.4, node_size=600)
```
